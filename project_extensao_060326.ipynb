## 3.4 Tradução de algumas variáveis categóricas
A variável experience_level representa uma hierarquia profissional. Para facilitar a comunicação com o público brasileiro, criaremos uma coluna traduzida.

Agora aplique um mapeamento para padronização (tradução da variável categórica) dos níveis de experiência.

Obs.: Mapeamento conceitual, significa associar um conceito de um sistema de representação a outro conceito semanticamente equivalente em outro sistema.

```
 nivel_de_experiencia = {
 "Entry - level ": "Jú nior ",
 "Mid - level ": " Pleno ",
 " Senior ": "Sê nior ",
 " Executive ": " Executivo "
 }

df_tratado [" experience_level "] = df_tratado [" experience_level "]. map (
nivel_de_experiencia )
```

### **Gráfico de barras**

- **Tipo de variável:** Ideal para variáveis categórica (qualitativa).
- **Permite visualizar:** frequência ou contagem de cada categoria.
- **Quando usar:** Para comparar categorias distintas.
- **Exemplo:**

### **Histograma**

- **Tipo de variável:** Ideal para variáveis numérica contínua.
- **Permite visualizar:** Distribuição dos dados em intervalos (classes).
- **Quando usar:** Para analisar concentração, assimetria e dispersão.
- **Exemplo:**

### **Gráfico de linhas**

- **Tipo de variável:** Ideal para variáveis numérica ao longo do tempo.
- **Permite visualizar:** Evolução, tendência e variação temporal.
- **Quando usar:** Para séries temporais.
- **Exemplo:**

### **Gráfico de pizza**

- **Tipo de variável:** Ideal para variáveis categórica.
- **Permite visualizar:** Proporção percentual entre categorias.
- **Quando usar:** Quando há poucas categorias e interesse em participação relativa.
- **Exemplo:**

### ***Boxplot***

- **Tipo de variável:** Ideal para variáveis numérica (comparada entre categorias).
- **Permite visualizar:** Mediana, quartis, dispersão e outliers.
- **Quando usar:** Para comparar distribuições entre grupos.
- **Exemplo:**

# 11. Gráfico de barras aplicado ao banco de dados

- Utilizado para analisar a variável categórica. Ex.: `job_category`.
- Justificativa: trata-se de uma variável qualitativa, ideal para comparação de frequências.
- Permite identificar qual categoria aparece com maior frequência.
- Facilita a comparação visual entre categorias.
- Não é adequado para analisar salários diretamente, pois salários são variáveis numéricas contínuas.

# 12. Histograma aplicado à distribuição salarial

- Aplicado à variável numérica contínua. Ex.: `salary_in_usd`.
- Permite observar concentração de salários em determinadas faixas.
- Indica possível assimetria (ex.: concentração em salários menores) e presença de valores extremos.
- É mais adequado que gráfico de barras porque trabalha com intervalos contínuos.

# 13. Gráfico de linhas e análise temporal

- Utilizado para analisar a mediana salarial ao longo dos anos (`work_year`).
- Permite identificar tendência de crescimento, queda ou estabilidade.
- É adequado para dados organizados cronologicamente.
- Fatores externos como inflação, mercado de tecnologia e demanda profissional podem influenciar os resultados.

# 14. Gráfico de pizza e proporções no banco de dados

- Aplicado à variável categórica `work_setting`.
- Permite visualizar a proporção de cada modelo de trabalho.
- Funciona bem quando há poucas categorias.
- Pode não ser adequado quando existem muitas categorias ou diferenças pequenas entre elas.

# 15. *Boxplot* e análise de dispersão dos salários

- Utilizado para comparar `salary_in_usd` entre níveis de `experience_level`.
- Mostra mediana, quartis e possíveis *outliers*.
- Permite verificar variação salarial entre níveis de experiência.
- Diferente do histograma, facilita a comparação entre grupos simultaneamente.

# 16. Gráfico de dispersão e relação entre variáveis

- Analisa a relação entre `work_year` e `salary_in_usd`.
- Permite identificar tendência de crescimento ou dispersão.
- Indica se há padrão ou ausência de relação.
- É essencial em análises exploratórias para verificar possíveis correlações.
